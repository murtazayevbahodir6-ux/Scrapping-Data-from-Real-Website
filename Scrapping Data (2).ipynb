# Scrapping Data 

### Countries in the world by population (2025)

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.worldometers.info/world-population/population-by-country/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
page
```

```python
print(soup)
```

##  Find and Find_all

```python
soup.find('div')
```

```python
soup.find_all('div')
```

```python
soup.find('table')
```

```python
soup.find_all('span')[0]
```

```python
world_title = soup.find_all('th')
```

```python
world_title
```

```python
word_table_title = [title.text.strip() for title in world_title]
print(word_table_title)
```

##  Data Frame

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = word_table_title)
df
```

```python
column_data = soup.find_all('tr')
column_data
```

```python
for row in column_data[1:]:    #     [1:] removes # column
    row_data = row.find_all('td')
    individual_raw_data = [row.text.strip()   for row in row_data]

    length = len(df)
    df.loc[length] = individual_raw_data

df
```
